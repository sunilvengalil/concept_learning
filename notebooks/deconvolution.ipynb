```python
import numpy as np
from  scipy.signal import correlate2d
from copy import copy
```

```python
a = np.asarray( [[1,2,3,4,5,6],[4,3,2,1,5,6],[3,2,1,4,5,6],[2,3,4,1,5,6],[1,2,3,4,5,6],[1,2,3,4,5,6]],
               dtype=np.float64 )
k = np.asarray( [[1.0,1.0,1.0],
                 [1.0,2.0,1.0],
                 [1.0,1.0,1.0]],
               dtype=np.float64)
print("Kernel")
print(k)
print("Image")
print(a)
```

```python
result = np.zeros(tuple([2 * s for s in a.shape]))
for i in range(0, a.shape[0]):
    for j in range(0, a.shape[1]):
        result[i:i+3,j:j+3] += a[i,j] * k
print(result)
```

```python
def insert_zeros(r,s):
    zero_inserted = np.zeros((s + 1) * r.shape[0])
    for i in range(r.shape[0]):
        zero_inserted[(s + 1) * i] = r[i]
    return zero_inserted
```

```python
def deconv(a, k, stride):
    result_1 = np.zeros(tuple([stride * s for s in a.shape]))
    for  i  in range(a.shape[0]):
        ai_zero_inserted = insert_zeros(a[i], stride- 1)
        result_1[stride * i,:] = ai_zero_inserted
    print(result_1)
    return correlate2d(result_1, k)
```

```python
result = deconv(a,k,1)
```

```python
def add_batch(t):
    result = np.expand_dims(t,0)
    return result

def add_batch_and_output_dims(t):
    result = np.expand_dims(t,0)
    return np.expand_dims(result,3)

    
```

```python
tf.reset_default_graph()

print(add_batch_and_output_dims(w).shape)

deconv = tf.nn.conv2d_transpose(tf.convert_to_tensor(add_batch(input_) ),
                                tf.convert_to_tensor(add_batch_and_output_dims(w) ),
                                output_shape=tuple(output_shape),
                                strides=[1, 2, 2, 1]
                               )
with tf.Session(config=tf.ConfigProto(allow_soft_placement=True)) as sess:
    deconv_result = sess.run(deconv)
    print(deconv_result)
```
